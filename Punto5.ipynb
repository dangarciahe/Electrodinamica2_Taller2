```python
import numpy as np
import matplotlib.pyplot as plt
```

---
# Punto 5. b)

```python
theta = np.linspace(0, 2*np.pi, 100)
r1 = np.sin(theta)**2

plt.polar(theta, r1)
plt.savefig("Graficas51.png")
plt.show()
```

---
# Punto 5. c)

```python
r2 = (1+np.cos(theta))**2

plt.polar(theta, r2)
plt.savefig("Graficas52.png")
plt.show()
```
