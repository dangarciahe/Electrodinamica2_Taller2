```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Wikipedia's values for water and vacuum
mur = 0.99992 # relative mu = mu/mu'
n1 = 1.0      # refraction index for vacum
n2 = 1.3330   # refraction index for water


## Angle in radians. Jackson pag. 305 
def r_perpendicular(i): #E'/E - Perpendicular
    i = np.radians(i)
    top = 2*n1*np.cos(i)
    bot = n1*np.cos(i) + mur*np.sqrt(n2**2 - (n1*np.sin(i))**2)
    return top/bot

def t_perpendicular(i): #E''/E - Perpendicular
    i = np.radians(i)
    top = n1*np.cos(i) - mur*np.sqrt(n2**2 - (n1*np.sin(i))**2)
    bot = n1*np.cos(i) + mur*np.sqrt(n2**2 - (n1*np.sin(i))**2)
    return top/bot


def r_parallel(i): #E'/E - Parallel
    i = np.radians(i)
    top = 2*n1*n2*np.cos(i)
    bot = mur*n2**2*np.cos(i) + n1*np.sqrt(n2**2 - (n1*np.sin(i))**2)
    return top/bot

def t_parallel(i): #E''/E - Parallel
    i = np.radians(i)
    top = mur*n2**2*np.cos(i) - n1*np.sqrt(n2**2 - (n1*np.sin(i))**2)
    bot = mur*n2**2*np.cos(i) + n1*np.sqrt(n2**2 - (n1*np.sin(i))**2)
    return top/bot

theta = np.linspace(0, 90, 100)

fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].plot(theta, r_parallel(theta), label="paralela")
ax[0].plot(theta, r_perpendicular(theta), label="perpendicular")
ax[1].plot(theta, t_parallel(theta), label="paralela")
ax[1].plot(theta, t_perpendicular(theta), label="perpendicular")


ax[0].set_ylabel(r"$r$")
ax[1].set_ylabel(r"$t$")

ax[0].set_title(r"Coeficiente $r$")
ax[1].set_title(r"Coeficiente $t$")

for ii in range (2):
    ax[ii].legend(title="Polarizacin:")
    ax[ii].grid()
    ax[ii].set_xticks(np.arange(0,100,10))
    ax[ii].set_xlabel(r"ngulo de incidencia, $i$")

plt.savefig("Graficas1.png")
plt.show()
```
